```python
page= requests.get('https://www.imdb.com/chart/top')
soup = BeautifulSoup(page.text, 'html.parser')
movie_list=[]
movie_names = []
movie_year=[0]*100
movie_rating=[]*100
j=0
for i in soup.find_all('td',class_="ratingColumn imdbRating"):
    if j<100:
        movie_rating.append(i.text.split('\n')[1])
        j=j+1
movie_rating
for i in range(100):
    content = str(soup.find_all('td', {'class':'titleColumn'})[i])
    name = re.findall ( '>(.*?)</a>', content)
    movie_names.insert(len(movie_names), name)
    year = str(soup.find_all('span', {'class':'secondaryInfo'})[i])
    movie_year[i] = int(re.findall(r"\(([0-9_]+)\)", year)[0])
movie_names
movie_year
```

```python
print(len(movie_names),len(movie_year),len(movie_rating))
```

```python
imdb_df=pd.DataFrame({'Movie Name': movie_names,'Year of Movie': movie_year,'Rating of Movie': movie_rating})
imdb_df
```

```python
#Top 10 Men's ODI Teams
page= requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
soup = BeautifulSoup(page.text, 'html.parser')
table = soup.find('table', {'class': 'table'})
i=1
rankings = {}
position = table.find('td', {'class': 'rankings-block__banner--pos'}).text.strip()
country_name = table.find('span', {'class': 'u-hide-phablet'}).text.strip()
matches = table.find('td', {'class': 'rankings-block__banner--matches'}).text.strip()
points = table.find('td', {'class': 'rankings-block__banner--points'}).text.strip()
rating = table.find('td', {'class': 'rankings-block__banner--rating u-text-right'}).text.strip()
rankings[country_name] = {'position': position,
                          'matches': matches,
                          'points': points,
                          'rating': rating}

for row in table.find_all('tr', {'class': 'table-body'}):
    if i<10:
        position = row.find('td', {'class': 'table-body__cell table-body__cell--position u-text-right'}).text.strip()
        i=i+1
        country_name = row.find('span', {'class': 'u-hide-phablet'}).text.strip()
        matches = row.find_all('td', {'class': 'table-body__cell u-center-text'})[0].text.strip()
        points = row.find_all('td', {'class': 'table-body__cell u-center-text'})[1].text.strip()
        rating = row.find('td', {'class': 'table-body__cell u-text-right rating'}).text.strip()
        rankings[country_name] = {'position': position,
                          'matches': matches,
                          'points': points,
                          'rating': rating}
rankings
```

```python
page= requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
soup = BeautifulSoup(page.text, 'html.parser')
Restaurant_name=[]
Cuisine=[]
Location=[]
Ratings=[]
Image_URL=[]
content=soup.find('div',class_="restnt-card-wrap-new")
for i in content.find_all('a',class_="restnt-name ellipsis"):
    Restaurant_name.append(i.text)
for i in content.find_all('div',class_="restnt-loc ellipsis"):
    Location.append(i.text)
for i in content.find_all('span',class_="double-line-ellipsis"):
    Cuisine.append(i.text.split('|')[1])
for i in content.find_all('div',class_="restnt-rating rating-4"):
    Ratings.append(i.text)
for i in content.find_all('img',class_="no-img"):
    Image_URL.append(i['data-src'])
#print(len(Restaurant_name),len(Cuisine),len(Location),len(Ratings),len(Image_URL))
dine_df=pd.DataFrame({'Restaurant_name':Restaurant_name,'Cuisine':Cuisine,'Location':Location,'Ratings':Ratings,'Image_URL':Image_URL})
dine_df
```
